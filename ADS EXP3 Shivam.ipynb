```python
import pandas as pd

# Assuming you have loaded your dataset into a DataFrame named 'startup_data'

startup_data = pd.read_csv('startup_data.csv')

# Define the intervals for each column
funding_intervals = pd.cut(startup_data['Funding Amount in $'], bins=5, labels=['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])
investor_intervals = pd.cut(startup_data['No. of Investors'], bins=5, labels=['Low', 'Medium-Low', 'Medium', 'Medium-High', 'High'])
year_intervals = pd.cut(startup_data['Starting Year'], bins=5, labels=['Early', 'Mid-Early', 'Mid', 'Mid-Late', 'Late'])

# Create a new DataFrame with interval columns
interval_data = pd.DataFrame({
    'Funding Intervals': funding_intervals,
    'Investor Intervals': investor_intervals,
    'Year Intervals': year_intervals
})

# Display the interval data
print("Interval Data:")
print(interval_data)

# Create DataFrame from the provided interval data
interval_data = pd.DataFrame({
    'Funding Intervals': ['Low', 'Low', 'Low', 'Low', 'Low', 'Low', 'Low', 'Low', 'Low', 'Low'],
    'Investor Intervals': ['Medium-Low', 'Medium', 'Low', 'Low', 'Medium-High', 'Medium', 'Low', 'Medium-High', 'Low', 'Low'],
    'Year Intervals': ['Late', 'Late', 'Mid-Late', 'Late', 'Late', 'Mid-Late', 'Late', 'Mid-Late', 'Late', 'Late']
})

# Calculate Pearson correlation coefficient for pairs of columns
column_pairs = [
    ['Funding Intervals', 'Investor Intervals'],
    ['Funding Intervals', 'Year Intervals'],
    ['Investor Intervals', 'Year Intervals']
]

for pair in column_pairs:
    column1, column2 = pair
    correlation_coefficient = interval_data[column1].astype('category').cat.codes.corr(interval_data[column2].astype('category').cat.codes, method='pearson')
    # Replace NaN with 0
    correlation_coefficient = correlation_coefficient if pd.notna(correlation_coefficient) else 0
    print(f"Pearson correlation coefficient between '{column1}' and '{column2}': {correlation_coefficient}")
```
