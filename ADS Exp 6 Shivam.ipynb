```python
import numpy as np
import pandas as pd
import math
import scipy.stats as stats

# Function to check for a symbol in an array
def check_for(sym, arr):
    i = [i for i in range(len(arr)) if arr[i] == sym]
    if i != []:
        return i[0]
    else:
        return None

def t_test(x, u, s, n, alpha):
    null_hypo = input("Enter Null hypothesis: ")
    null_hypo = null_hypo.split(' ')
    alternate_hypo = [i for i in null_hypo]
    test_type = ''

    # Check hypothesis
    if check_for('>=', null_hypo):
        alternate_hypo[check_for('>=', null_hypo)] = '<'
        test_type = 'Left Tailed Test'
    elif check_for(' <= ', null_hypo):
        alternate_hypo[check_for(' <= ', null_hypo)] = '>'
        test_type = 'Right Tailed Test'
    elif check_for('>', null_hypo):
        alternate_hypo[check_for('>', null_hypo)] = ' <= '
        test_type = 'Left Tailed Test'
    elif check_for('<', null_hypo):
        alternate_hypo[check_for('<', null_hypo)] = '>='
        test_type = 'Right Tailed Test'
    elif check_for('=', null_hypo):
        alternate_hypo[check_for('=', null_hypo)] = ' != '
        test_type = 'Two Tailed Test'
    else:
        print("Failed!")
        return

    # Calculate degrees of freedom
    df = n - 1

    # Calculate t-Score, P-value, Critical Value
    t_score = (x - u) / (s / math.sqrt(n))

    if test_type == 'Left Tailed Test':
        critical_value = stats.t.ppf(alpha, df)
    elif test_type == 'Right Tailed Test':
        critical_value = stats.t.ppf(1 - alpha, df)
    else:
        critical_value = stats.t.ppf(1 - alpha, df)

    if test_type == 'Left Tailed Test':
        p_value = stats.t.cdf(t_score, df)
    elif test_type == 'Right Tailed Test':
        p_value = stats.t.cdf(t_score, df)
    else:
        p_value = 2 * (1 - stats.t.cdf(abs(t_score), df))

    # Print the results
    print('Null (Ho) : ', " ".join(null_hypo))
    print('Alternate (Ha) : ', " ".join(alternate_hypo))
    print('Test type : ' + test_type)
    print('T-Score : ', end='')
    print(t_score)
    print('P-value : ', end='')
    print(p_value)
    print('Critical Value : ', end='')
    print(critical_value)

from google.colab import drive
drive.mount('/content/drive')

# Load your CSV data into a DataFrame
data = pd.read_csv("/content/drive/MyDrive/startup_data.csv")
df = pd.read_csv("/content/drive/MyDrive/startup_data.csv")

# Sample data from your DataFrame
sample = df.sample(n=100)

# Calculate mean, standard deviation, and count for the sample
u = df['Funding Amount in $'].mean()
x = sample['Funding Amount in $'].mean()
s = sample['Funding Amount in $'].std()
n = sample['Funding Amount in $'].count()
alpha = 0.05  # Significance level

# Perform t-test
t_test(x, u, s, n, alpha)

import numpy as np
import scipy.stats as stats

# Define the sample data and parameters
X = 22
u = 24
S = 4.8
n = 27
alpha = 0.01

# Calculate the t-score
t_score = (X - u) / (S / np.sqrt(n))

# Calculate the p-value
p_value = 1 - stats.t.cdf(t_score, df=n-1)

# Calculate the critical value
critical_value = stats.t.ppf(1 - alpha, df=n-1)

# Print the results
print("Null Hypothesis (Ho): u <= 9")
print("Alternate Hypothesis (Ha): u > 9")
print("Test type: Right-Tailed Test")
print("T-Score:", t_score)
print("P-value:", p_value)
print("Critical Value:", critical_value)

# Determine if null hypothesis should be rejected
if p_value < alpha:
    print("\nNull Hypothesis Ho is rejected!")
    print("Alternate Hypothesis Ha is accepted!")
else:
    print("\nNull Hypothesis Ho is accepted.")
    print("No significant evidence to reject Ho.")
```
